```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
```

```python
f105w_data = fits.getdata('f105w.fits')
f125w_data = fits.getdata('f125w.fits')
f160w_data = fits.getdata('f160w.fits')
```

```python
rgb_image = np.stack([f160w_data, f125w_data, f105w_data], axis=-1)
```

```python
plt.imshow(rgb_image)
plt.axis('off')  # Hide axes
plt.title("3-Color False Image of HUDF")
plt.savefig('hubble_3color.png', dpi=300)
plt.show()
plt.savefig('3CFI')
```
